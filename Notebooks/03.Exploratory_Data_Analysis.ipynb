```python
import pandas as pd
listings=pd.read_csv("clean_listings.csv")
```

```python
#top neighbourhoods by revenue
import matplotlib.pyplot as plt
top_areas=listings.groupby("neighbourhood_cleansed")["estimated_revenue_l365d"].mean().sort_values(ascending=False).head(15)
top_areas.plot(kind="bar",figsize=(12,6))
plt.title("Top Neighbourhoods by Avg Revenue")
plt.show()
```

```python
#Revenue by room type
listings.groupby("room_type")["estimated_revenue_l365d"].mean().sort_values(ascending=False).plot(kind="bar")
plt.title("Room Types by Avg Revenue")
```

```python
#Revenue by property type
top_property=listings.groupby("property_type")["estimated_revenue_l365d"].mean().sort_values(ascending=False).head(15)
top_property.plot(kind="bar")
plt.title("Property Type by Avg Revenue")
plt.show()
```

```python
#Superhost vs non-superhost
listings.groupby("host_is_superhost")["estimated_revenue_l365d"].mean().plot(kind="bar")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    listings["estimated_occupancy_l365d"],
    listings["estimated_revenue_l365d"],
    alpha=0.3
)
plt.xlabel("Occupancy")
plt.ylabel("Revenue")
plt.title("Occupancy vs Revenue")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    listings["review_scores_rating"],
    listings["estimated_revenue_l365d"],
    alpha=0.3
)
plt.xlabel("Rating")
plt.ylabel("Revenue")
plt.title("Rating vs Revenue")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    listings["amenity_count"],
    listings["estimated_revenue_l365d"],
    alpha=0.3
)
plt.xlabel("Amenity Count")
plt.ylabel("Revenue")
plt.title("Amenities vs Revenue")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    listings["host_experience_years"],
    listings["estimated_revenue_l365d"],
    alpha=0.3
)
plt.xlabel("Host Experience (Years)")
plt.ylabel("Revenue")
plt.title("Host experience vs Revenue")
plt.show()
```

```python
listings["host_response_rate"]=(listings["host_response_rate"].str.replace("%","") .astype(float))
listings["host_acceptance_rate"]=(listings["host_acceptance_rate"].str.replace("%","") .astype(float))
```

```python
corr_cols=[
    "price",
    "accommodates",
    "bedrooms",
    "bathrooms",
    "amenity_count",
    "host_experience_years",
    "review_scores_rating",
    "estimated_occupancy_l365d",
    "estimated_revenue_l365d"
]
corr=listings[corr_cols].corr()
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True)
plt.title("Correlation Matrix")
plt.show()
```
